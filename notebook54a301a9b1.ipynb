```python
x = torch.tensor([[1.6], [1.7], [1.8], [0]], dtype=torch.float32)

y_true = torch.tensor([[60.0], [65.0], [70.0],[-20]], dtype=torch.float32)

layer = nn.Linear(1, 1)

#example using height and weight
#given a height in meters (x) predict the weight in kg (y) using one linear layer

y_hat = layer(x)

print("Predictions (y_hat):", y_hat)
print("True values (y):", y_true)
loss_func = nn.MSELoss()
loss = loss_func(y_hat, y_true)
print("loss:", loss)
```

loss function is mean squared error: $$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

$$
\text{MSE} = \frac{1}{3} \left[ ( y_1 - \hat{y}_1 )^2 + ( y_2 - \hat{y}_2 )^2 + ( y_3 - \hat{y}_3 )^2 \right]
$$

```python
optimizer = torch.optim.SGD(layer.parameters(), lr=0.01)

#loop 100 times
for epoch in range(1000):
    optimizer.zero_grad()  
    y_hat = layer(x)       # Forward pass (compute y_hat)
    loss = loss_func(y_hat, y_true)  # Compute loss
    loss.backward()        # Backpropagation (compute gradients)
    optimizer.step()       # Update parameters (m and b)

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch + 1}, Loss: {loss.item():.2f}")

print("\nTrained weight (m):", layer.weight)
print("Trained bias (b):", layer.bias)
```
